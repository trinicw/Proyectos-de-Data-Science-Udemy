```python
import pandas as pd

df = pd.read_csv('Top-Películas.csv')
df
```

Ordenación de DataFrames

```python
df_ordenado = df.sort_values(by= 'rating')
df_ordenado.head(10)
```

```python
df_ordenado = df.sort_values(by= 'rating', ascending=False)
df_ordenado.head(10)
```

```python
df_ordenado = df.sort_values(by= ['rating', 'recaudación(M)'], ascending=False)
df_ordenado.head(10)
```

Agrupacion de DF : reuinir datos basandonos en categorias o valores comunes

```python
# Ejemplo : Agrupar las peliculas por genero y calcular su promedio
df_agrupado = df.groupby('género')['rating'].mean()  #De esta manera estoy calculando el raiting de cada uno de esos grupos
df_agrupado
```

```python
# Obtener el total recaudado por año
df_agrupado = df.groupby('año')['recaudación(M)'].sum()
df_agrupado.sort_values(ascending=False )
```
